For my replication study, I used the article “Murderous, Unwarrantable, and Very Cold”: Mapping the Rise of Extralegal Collective Killing in the United States, 1783-1865” by Patrick Hoehne, published in 2022. This project investigates the rise of killing as a form of extralegal violence, and how such incidents served as a form of social communication. Hoehne’s primary graphs depict the incidents of extralegal killing from 1783 to 1865, the recorded fatalities resulting from extralegal violence between the same years, the U.S. population from 1790 to 1860, and a bar graph showing the number of instances of lethal extralegal collective violence per 100,000 U.S. Residents between the years 1790 and 1860. As Hoehne wrote, “This surge in extralegal collective killings must be read alongside U.S. population growth, which also experienced a dramatic rise during the period surveyed in this article.” However, I noticed that while a bar graph demonstrated the rate of extralegal instances, there was no depiction of the rate of actual killings. Therefore, instead of replicating a graph already utilized by this project, I created a new graph to show this rate. To begin this process, I collected numerical data from the article and organized it in a spreadsheet. The dataset included the total number of killings within each decade and the corresponding U.S. population between 1790 and 1860. Because the killings were grouped by decade, I first calculated the average number of killings per year by dividing the total number of killings in each decade by ten. Next, I adjusted these values to account for population growth. To do this, I estimated the average population for each decade by taking the midpoint between the two census values that bracketed the decade. For example, to estimate the population for 1790–1800, I averaged the populations recorded in the 1790 and 1800 censuses. I then calculated the rate of killings per 100,000 residents using the formula: (average annual killings ÷ average population) × 100,000. This calculation allowed the killings to be standardized relative to population size, making it possible to compare levels of violence across decades. After completing these calculations, I entered the final values into a spreadsheet with columns for decade, average annual killings, estimated population, and killings per 100,000 residents. I then saved the spreadsheet as a comma-separated values (.csv) file, which allowed it to be easily imported into my Colab program. Using Python, I generated a graph that visualized the rate of killings per 100,000 residents across decades. It is important to note that because my graph measured the number of deaths rather than the number of incidents, the values were often higher than those in Hoehne’s graph, since a single extralegal event could result in multiple fatalities. Although this graph provides a useful way to depict violence in relation to population change, the data does have limitations. Historical records of extralegal violence may be incomplete due to bias or inconsistent recordkeeping in earlier decades. This limitation suggests that the data likely underrepresents the true number of incidents and fatalities.
